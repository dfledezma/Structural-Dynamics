The frequency response of a SDOF model with viscous damping is analysed in this notebook. The equation of motion is given as:

$\quad m \ddot{x} + c \dot{x} + kx = F(t)$

Using the complex exponential approach, the input force, the displacement response, and its derivatives can be written as:

Assume a solution for simple harmonic motion displacement and its derivatives using complex exponential notation in the form:

$F(t)=Fe^{i\omega t}$

$x(t)=X e^{i\omega t}$

$\dot{x}(t)=i\omega_n X e^{i\omega t}$

$\ddot{x}(t)=-\omega_n^2 X e^{i\omega t}$

As the system is linear we can assume that for a harmonic input, the response will be of the same form. Substituting the input forceand the assumed response the equation can be written as:

$-\omega^2 m X + i \omega c X + kX = F$

The FRF, i.e. the output/input ratio as a function of frequency is then given by :

$\frac{X}{F}=\frac{1}{k -\omega^2 m + i \omega c}$

This is a complex function. It is also writthen in nondimensional form as

$\frac{X}{\Delta}=\frac{1}{1-\Omega^2+i2\zeta \Omega}$

Where $\Delta=\frac{F}{k}$ is the static deflection, and $\Omega = \frac{\omega}{\omega_n}$ is the frequency ratio. Expressing the FRF by its magnitude and phase angle:

$\left|  \frac{X}{\Delta} \right| =\frac{1}{\sqrt {(1-\Omega^2)^2+(2 \zeta \Omega})^2}$

$\phi = tan^{-1} =\frac{2 \zeta \Omega}{ 1-\Omega^2}$




```python
import numpy as np    # numpy is the main module for numeric calculations, which will be called as np

%matplotlib inline

# Import the plotting module 
import matplotlib.pyplot as plt
```

```python
# Set up parameters
wr = np.linspace(0,10,1000)          # Frequency ratio range for FRF plot

zeta = 0.01
FRF_mag_01 = np.abs(1/(1-wr**2 + 1j*2*zeta*wr))
FRF_phase_01 = np.angle(1/(1-wr**2 + 1j*2*zeta*wr))* 180/np.pi

zeta = 0.05
FRF_mag_05 = np.abs(1/(1-wr**2 + 1j*2*zeta*wr))
FRF_phase_05 = np.angle(1/(1-wr**2 + 1j*2*zeta*wr))* 180/np.pi

zeta = 0.1
FRF_mag_1 = np.abs(1/(1-wr**2 + 1j*2*zeta*wr))
FRF_phase_1 = np.angle(1/(1-wr**2 + 1j*2*zeta*wr))* 180/np.pi

zeta = 0.3
FRF_mag_3 = np.abs(1/(1-wr**2 + 1j*2*zeta*wr))
FRF_phase_3 = np.angle(1/(1-wr**2 + 1j*2*zeta*wr))* 180/np.pi

# Plot the magnitude

# Change the size of plot
fig = plt.figure(figsize=(9,6))

# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Plot grid can be activated and modified
ax.grid(True,linestyle=':',color='0.5')
ax.set_axisbelow(True)

# Define axis labels
plt.xlabel('Frequency ratio $\Omega$',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('FRF Magnitude',family='serif',fontsize=18,weight='bold',labelpad=10)


plt.plot(wr, FRF_mag_01, linewidth=2, linestyle = '-', label=r'$\zeta$ = 0.01')

plt.plot(wr, FRF_mag_05, linewidth=2, linestyle = '-.', label=r'$\zeta$ = 0.05')

plt.plot(wr, FRF_mag_1, linewidth=2, linestyle = '--', label=r'$\zeta$ = 0.1')

plt.plot(wr, FRF_mag_3, linewidth=2, linestyle = ':', label=r'$\zeta$ = 0.3')

plt.xlim(0,3)
plt.ylim(0,15)

leg = plt.legend(loc='upper right', fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext,family='Serif',fontsize=16)
```

```python
# Plot the phase

# Change the size of plot
fig = plt.figure(figsize=(9,6))

# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Plot grid can be activated and modified
ax.grid(True,linestyle=':',color='0.5')
ax.set_axisbelow(True)

# Define axis labels
plt.xlabel('Frequency ratio $\Omega$',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('FRF Phase angle',family='serif',fontsize=18,weight='bold',labelpad=10)

plt.plot(wr, FRF_phase_01, linewidth=2, linestyle = '-', label=r'$\zeta$ = 0.01')

plt.plot(wr, FRF_phase_05, linewidth=2, linestyle = '-.', label=r'$\zeta$ = 0.05')

plt.plot(wr, FRF_phase_1, linewidth=2, linestyle = '--', label=r'$\zeta$ = 0.1')

plt.plot(wr, FRF_phase_3, linewidth=2, linestyle = ':', label=r'$\zeta$ = 0.3')

plt.xlim(0,3)

leg = plt.legend(loc='upper right', fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext,family='Serif',fontsize=16)
```

FRF is sometimes best represented using logarmitmic axis (or dB) 

```python
# Plot the magnitude using loglog axis

# Change the size of plot
fig = plt.figure(figsize=(9,6))

# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Plot grid can be activated and modified
ax.grid(True,linestyle=':',color='0.5')
ax.set_axisbelow(True)

# Define axis labels
plt.xlabel('Frequency ratio $\Omega$',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('FRF Magnitude',family='serif',fontsize=18,weight='bold',labelpad=10)


plt.loglog(wr, FRF_mag_01, linewidth=2, linestyle = '-', label=r'$\zeta$ = 0.01')

plt.loglog(wr, FRF_mag_05, linewidth=2, linestyle = '-.', label=r'$\zeta$ = 0.05')

plt.loglog(wr, FRF_mag_1, linewidth=2, linestyle = '--', label=r'$\zeta$ = 0.1')

plt.loglog(wr, FRF_mag_3, linewidth=2, linestyle = ':', label=r'$\zeta$ = 0.3')

plt.xlim(0.1,10)
plt.ylim(0.01,100)

leg = plt.legend(loc='upper right', fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext,family='Serif',fontsize=16)
```

The ratio between displacement response and input force is called "Receptance" and  its inverse is the dynamic stifness. There are other forms of the FRF depending on the input/output variables. By multiplying by the factor $ i \omega$ the one can go from displacement to velocity. This function is called mobility, and its inverse represents mechanical impedance:

$\frac{\dot{X}}{F} =  i \omega \left(\frac{1}{k -\omega^2 m + i \omega c}\right)$

This function is called mobility, and its inverse represents mechanical impedance. By multiplying by $ i \omega$ again, the resulting acceleration/fore ratio is called accelerance and it inverse is the apparent mass. 

$\frac{\ddot{X}}{F} =  -\omega^2 \left(\frac{1}{k -\omega^2 m + i \omega c}\right)$



```python
# Calculate mobility

zeta = 0.01
mob_mag_01 = np.abs(1j*wr*(1/(1-wr**2 + 1j*2*zeta*wr)))

# Calculate accelerance
acc_mag_01 = np.abs(-wr**2*(1/(1-wr**2 + 1j*2*zeta*wr)))

# Plot the FRFs

# Change the size of plot
fig = plt.figure(figsize=(9,6))

ax1 = plt.subplot(211)
ax1.loglog(wr, mob_mag_01, linewidth=2, linestyle = '-')
ax1.set_title('Mobility' ,fontsize=16)

# Define axis labels
plt.xlabel('Frequency ratio $\Omega$',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('Mobility Magnitude',family='serif',fontsize=18,weight='bold',labelpad=10)
# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.01,left=0.17,top=2,right=0.96)


ax2 = plt.subplot(212, sharex=ax1)
ax2.loglog(wr, acc_mag_01, linewidth=2, linestyle = '-')
ax2.set_title('Accelerance', fontsize=16)

# Define axis labels
plt.xlabel('Frequency ratio $\Omega$',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('Accelerance Magnitude',family='serif',fontsize=18,weight='bold',labelpad=10)

# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.01,left=0.17,top=2,right=0.96)


plt.xlim(0.1,10)
plt.ylim(0.01,100)


plt.show()
```
