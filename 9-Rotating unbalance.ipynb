The frequency response of a SDOF model due to unbalance is analysed in this notebook. Consider a single degree of freedom system with a rotating element of frequency $\omega$ and an idealised unbalance mass $m_u$ that generates an unbalance harmonic force $F(t)$.



The equation of motion is given as:

$\quad m \ddot{x} + c \dot{x} + kx = F_u(t)$

Considering that the unbalance mass has a harmonic displacement given by the distance $e$ that represents the excentricity, i.e. the distance from the center of rotation to the position of the idealised unbalance mass $m_u$, given by $e(t) = e sin(\omega t)$, the magnitude of the unbalance mass acceleration is then $e \omega^2$. As a result, the unbalance force $F_u(t)$ is given by the product of the unbalance mass $m_u$ and its acceleration, i.e. $F_u=m_u e \omega^2$,  The steady state response of a SDOF system under an unbalance force is given by:

$X=\frac{m_u \omega^2}{k -\omega^2 m + i \omega c}$

Also expressed in nondimensional form as:

$\frac{M_t X}{m_u e}=\frac{\Omega^2}{1-\Omega^2+i2\zeta \Omega}$

$\frac{M_t X}{m_u e}$ is the amplification ratio due to unbalance, and $\Omega = \frac{\omega}{\omega_n}$ is the frequency ratio.



```python
import numpy as np    # numpy is the main module for numeric calculations, which will be called as np

%matplotlib inline

# Import the plotting module 
import matplotlib.pyplot as plt
```

```python
# Set up parameters
wr = np.linspace(0,10,1000)          # Frequency ratio range for FRF plot

zeta = 0.01
FRF_unb_01 = np.abs(wr**2/(1-wr**2 + 1j*2*zeta*wr))

zeta = 0.05
FRF_unb_05 = np.abs(wr**2/(1-wr**2 + 1j*2*zeta*wr))

zeta = 0.1
FRF_unb_1 = np.abs(wr**2/(1-wr**2 + 1j*2*zeta*wr))

zeta = 0.3
FRF_unb_3 = np.abs(wr**2/(1-wr**2 + 1j*2*zeta*wr))

# Plot the magnitude

# Change the size of plot
fig = plt.figure(figsize=(9,6))

# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Plot grid can be activated and modified
ax.grid(True,linestyle=':',color='0.5')
ax.set_axisbelow(True)

# Define axis labels
plt.xlabel('Frequency ratio $\Omega$',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('Unbalance amplification',family='serif',fontsize=18,weight='bold',labelpad=10)


plt.plot(wr, FRF_unb_01, linewidth=2, linestyle = '-', label=r'$\zeta$ = 0.01')

plt.plot(wr, FRF_unb_05, linewidth=2, linestyle = '-.', label=r'$\zeta$ = 0.05')

plt.plot(wr, FRF_unb_1, linewidth=2, linestyle = '--', label=r'$\zeta$ = 0.1')

plt.plot(wr, FRF_unb_3, linewidth=2, linestyle = ':', label=r'$\zeta$ = 0.3')

plt.xlim(0,3)
plt.ylim(0,15)

leg = plt.legend(loc='upper right', fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext,family='Serif',fontsize=16)
```
