Plots. matplotlib is used for plotting and generate print quality plots

```python
# Figures will be presented inline
%matplotlib inline

# Import matplotlib abd call it as plt 
import matplotlib.pyplot as plt
```

Example of a sine plot.

```python
t = np.linspace(0, 10, 1000)  # Define the time vector using linspace (start,end,number of points)
# Define amplitude and frequency
Amp1=1
omega=1
Amp2=1.5
y = Amp1*np.sin(omega * t) 
y2 = Amp2*np.cos(omega * t)
```

Plot and compare the functions. Are they easy to understand?

```python
# Call matplot lib as plt in the plot command
# This means "use the matplotlib version of this command"
# Note the use of commands to modify line style, color and weight
plt.plot(t, y, color="blue", linewidth=1.0)
plt.plot(t, y2, color="green", linewidth=1.0)
# Label X and Y axes 
# Note that labelpad is the spacing in points from the axes bounding box including ticks and tick labels.
plt.xlabel(r'Time (s)', fontsize=18, labelpad=5)
plt.ylabel(r'$A\sin(f\pi t)$',fontsize=18, labelpad=8)
```

LetÂ´s try to make the plot better looking

```python
# Change the size of plot
fig = plt.figure(figsize=(9,6))

# Get current axes using gca
ax = plt.gca()

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=16)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=16)

# Tweak spacing to prevent clipping of labels
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Box frame can be modified, i.e. remove top and right borders (spines), or move them to the origin 
# ax.spines['right'].set_color('none')
# ax.spines['top'].set_color('none')

# Plot grid can be activated and modified
ax.grid(True,linestyle=':',color='0.5')
ax.set_axisbelow(True)

# Define axis labels
plt.xlabel('Time (s)',family='serif',fontsize=18,weight='bold',labelpad=5)
plt.ylabel('Y (units)',family='serif',fontsize=18,weight='bold',labelpad=10)

# Plot with different line styles and add labels

plt.plot(t, y, color="blue", linewidth=1.0, linestyle='-', label=r' $A_1 sin(\omega t)$')
plt.plot(t, y2, color="green", linewidth=1.0, linestyle='--', label=r' $A_2 cos(\omega t)$')

# Set axis limits
plt.xlim(0,10)
plt.ylim(-2.5,2.5)

# Add the legend calling the command, specify position and columns
leg = plt.legend(loc='upper right', ncol = 2, fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext,family='serif',fontsize=16)

# Adjust the page layout to fill the page using the new tight_layout command
plt.tight_layout(pad=0.5)

# Export the figure es pdf
plt.savefig('two_sines.pdf')

# You can also add annotations if desired.
# Mark maximum peak amplitude of sine function
tp=omega*1/2*np.pi

plt.plot([tp, tp], [0, Amp1*np.sin(omega*tp)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([tp, ], [Amp1*np.sin(omega*tp), ], 50, color='blue')


plt.annotate(r'$sin(\pi/2)=A_{peak}$',
             xy=(tp, Amp1*np.sin(omega*tp)), xycoords='data',
             xytext=(-80, 50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc,rad=1"))

# Mark maximum peak amplitude of cosine function
tp=omega*2*np.pi

plt.plot([tp, tp], [0, Amp2*np.cos(omega*tp)], color='red', linewidth=2.5, linestyle="--")

plt.annotate(r'$cos(2\pi)=A_{peak}$',
             xy=(tp, Amp2*np.cos(omega*tp)), xycoords='data',
             xytext=(-100, 20), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc,rad=1"))

plt.scatter([tp, ], [Amp2*np.cos(omega*tp), ], color='red')
```
